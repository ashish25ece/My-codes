```python
#segregate the independent and dependent variable

X= sonar_data.iloc[:,:-1]
y= sonar_data.iloc[:,-1]
```

```python
# Since the target variable is a categorical variable change it into numerical value 
# using LableEncoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#Scaling the data

from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()
X_scaled= scaler.fit_transform(X)
```

```python
X=pd.DataFrame(X_scaled,columns=X.columns)
X.shape
```

```python
# Create test and train data set

from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

## Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logreg= LogisticRegression()
logreg.fit(train_X,train_y)
```

```python
#predicing the output on the test set and checking the classification report

lr_predicted= logreg.predict(test_X)

print(classification_report(test_y,lr_predicted))
```

```python
lr_predicted
```

```python
#check the accuracy of the Logisctic regresion model

from sklearn.metrics import accuracy_score
lr_score=accuracy_score(test_y,lr_predicted)
print(lr_score)

from sklearn.metrics import confusion_matrix
print(confusion_matrix(test_y,lr_predicted))
```

```python
#checking the f1 score of the LR model

from sklearn.metrics import f1_score

lr_f1score= f1_score(test_y,lr_predicted)
lr_f1score
```

## Implementing KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier as KNN

knn=KNN()
knn.fit(train_X,train_y)
```

```python
knn_predicted=knn.predict(test_X)
knn_score=accuracy_score(test_y,knn_predicted)
#knn_score= accuracy_score(test_y,knn_predicted)
print(knn_score)
```

```python
#Defining fucntion to find the best value of k for thr KNN classifier

def Elbow(K):
    #initialiizng the empty list
    test_error=[]
    
    #training model for every value of k
    for i in k:
        knn=KNN(n_neighbors=i)
        knn.fit(train_X,train_y)
        knn_predicted=knn.predict(test_X)
        
        knn_score=accuracy_score(test_y,knn_predicted)
        error_score=1-knn_score
        test_error.append(error_score)
    return test_error
```

```python
#Implementing and calling the fucntion to find the best value of K

k=range(1,20,2)

test= Elbow(k)
#test
#plt.figure(figsize=(10,10))
plt.plot(k,test)
```

```python
#Predciting the output with the best value of k

knn=KNN(n_neighbors=7)
knn.fit(train_X,train_y)

knn_predicted=knn.predict(test_X)
knn_score=accuracy_score(test_y,knn_predicted)
#knn_score= accuracy_score(test_y,knn_predicted)
print(knn_score)
```

```python
#checking the f1 score of the knn model

knn_f1score= f1_score(test_y,knn_predicted)
knn_f1score
```

## Implementing Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(train_X,train_y)
```

```python
nb_predicted = nb.predict(test_X)
nb_score = nb.score(test_X,test_y)
nb_score
```

```python
#checking the f1 score of the NB model

nb_f1score= f1_score(test_y,nb_predicted)
nb_f1score
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=10)
dt.fit(train_X,train_y)
```

```python
dt_score=dt.score(test_X,test_y)
dt_score
```

```python
#Defining fucntion to find the best value of random state for thr KNN classifier

def decisiontree_state(K):
    #initialiizng the empty list
    test_error=[]
    
    #training model for every value of k
    for i in k:
        dt = DecisionTreeClassifier(random_state=i)
        dt.fit(train_X,train_y)
        dt_predicted=dt.predict(test_X)
        
        dt_score=accuracy_score(test_y,dt_predicted)
        error_score=1-dt_score
        test_error.append(error_score)
    return test_error
```

```python
random_value=(1,20,2)
test_value=decisiontree_state(random_value)

test_value
```

```python
#Predciting the output with the best value of random state

dt = DecisionTreeClassifier(random_state=5)
dt.fit(train_X,train_y)
dt_predicted= dt.predict(test_X)
dt_score=dt.score(test_X,test_y)
dt_score
```

```python
#checking the f1 score of the DT model

dt_f1score= f1_score(test_y,dt_predicted)
dt_f1score
```

```python
# Importing the libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data_set= pd.read_csv(r"C:\Users\Ash\Desktop\Data Trained\Projects\Project 1- Titanic/train.csv")
```

```python
data_set.head()
```

```python
data_set
```

```python
#check if there is any missing value
data_set.isnull().sum()
```

```python
#Since feature cabin has more missig values, we can drop this feature
data_set = data_set.drop(['Cabin'], axis=1)
```

```python
#Since feature Age has missig values, we can fill it with mean

data_set['Age'] = data_set['Age'].fillna((data_set['Age'].mean()))
```

```python
data_set
```

```python
# Convert the categorical values into numerical values and concatenate with the train set

data_set = pd.get_dummies(data_set, columns=['Sex', 'Embarked'])
```

```python
data_set.columns
```

```python
data_set.drop(['PassengerId','Name','Ticket'],axis=1, inplace=True)
```

```python
#check the correlation between the features
data_set.corr()
```

```python
#segregate the independent and dependent variable

X = data_set.drop('Survived',axis =1)
y= data_set['Survived']
```

```python
#Scaling the data

from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()
X_scaled= scaler.fit_transform(X)
```

```python
X=pd.DataFrame(X_scaled,columns=X.columns)
X
```

```python
#Splitting the train and test data set

from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y= train_test_split(X,y,test_size=0.33,random_state=42,stratify=y)
```

## Implementing the Logistic Regression 

```python
# Implementing Logistic Regression

from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()
logreg.fit(train_X,train_y)
```

```python
train_predict= logreg.predict(train_X)
```

```python
# checking the F1 score and others on the training set

from sklearn.metrics import classification_report

print(classification_report(train_predict, train_y))
```

```python
# creating the consfusion matrix on the train set

from sklearn.metrics import confusion_matrix

print(confusion_matrix(train_y,train_predict))
```

```python
from sklearn import metrics
print(metrics.accuracy_score(train_y, train_predict))
```

```python
#Predcitng the output on the test set

test_predict= logreg.predict(test_X)

print(classification_report(test_predict, test_y))
```

```python
# creating the consfusion matrix  on the train set
print(confusion_matrix(test_y,test_predict))
```

```python
from sklearn import metrics
print(metrics.accuracy_score(test_y, test_predict))
```

```python
#scaling the data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sales_data_scaled = scaler.fit_transform(sales_data)

sales_data = pd.DataFrame(sales_data_scaled,columns=sales_data.columns)
sales_data
```

```python
#segregate the dependent and independent variable

X = sales_data.iloc[:,:-1]
y = sales_data.iloc[:,-1]
```

```python
#splitting the dataset into train and test data set

from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.33,random_state= 42)
```

## Implementing the Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_X,train_y)
```

```python
lr_predicted = lr.predict(test_X)

from sklearn.metrics import mean_squared_error

lr_error = mean_squared_error(test_y,lr_predicted)
lr_error
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr_score = lr.score(test_X,test_y)
lr_score
```
